Pregunta: Qué actores han trabajado en más de dos películas de drama

```python
#SELECT COUNT(*), Actor_Movie.act_id FROM movie INNER JOIN Actor_Movie ON movie.mov_id = Actor_Movie.mov_id WHERE gen_id = 1 GROUP BY act_id HAVING COUNT(*) > 2
df_join = pd.merge(movie, ma, how = 'inner', left_on = 'mov_id', right_on = 'mov_id')  
df_join = df_join[df_join.genero == 1007]
df_join
```

```python
df_join = df_join.groupby('act_id').agg({'act_id' : 'count'})
```

```python
df_join = df_join.rename(columns = {'act_id' : 'count'})
```

```python
df_join[df_join['count'] > 2]
```
